## Compute margin of error, confidence interval, and p-value  

```python
propBSounding = (blackGotCall + blackNoCall)/sampleN
propWhiteSounding = (whiteGotCall + whiteNoCall)/sampleN

propGotCall = gotCall/sampleN
propNoCall = noCall/sampleN

propBlackGotCall = blackGotCall / sampleB
propBlackNoCall = blackNoCall / sampleB

propWhiteGotCall = whiteGotCall/sampleW
propWhiteNoCall = whiteNoCall/sampleW

print("Proportion of Black Sounding {:.5f}".format(propBSounding))
print("Proportion of White Sounding {:.5f}".format(propWhiteSounding))

print("Proportion of Resumes got call back {:.5f}".format(propGotCall))
print("Proportion of Resumes that did not get call back {:.5f}".format(propNoCall))

print("Proportion of callbacks for black-sounding names {:.5f}".format(propBlackGotCall))
print("Proportion of callbacks for white-sounding names {:.5f}".format(propWhiteGotCall))
print("Proportion of zero callbacks for black-sounding names {:.5f}".format(propBlackNoCall))
print("Proportion of zero callbacks for white-sounding names {:.5f}".format(propWhiteNoCall))
```

```python
# We are interested in p_w-p_b
propDiffWhiteBlack = abs(propWhiteGotCall - propBlackGotCall)
print('Observed difference in proportion of call backs: {:.5}'.format(propDiffWhiteBlack))

sampleWhiteVar = (propWhiteGotCall * (1-propWhiteGotCall))/sampleW
sampleBlackVar = (propBlackGotCall * (1-propBlackGotCall))/sampleB


stdErrorWhite = np.sqrt(sampleWhiteVar)
stdErrorBlack = np.sqrt(sampleBlackVar)

stdError = np.sqrt(sampleWhiteVar + sampleBlackVar)
print('The standard error: {:.5}'.format(stdError))
```

```python
zCritical1Tail = stats.norm.ppf(0.95)
print('Z Critical at 1 tail 95% confidence: {:.5}'.format(zCritical1Tail))

marginOfError = zCritical1Tail * stdError
print('The margin of error on our observation at 95% confidence (1 tail) is {:.4}'.format(marginOfError))

print('\nThe difference in proportion of people reciving call backs: {:.4} +/- {:.4}'.format(
    propDiffWhiteBlack,marginOfError))

[ciu,cil] = zCritical1Tail*stdError*np.array([-1,1])
ciUpper = propDiffWhiteBlack + ciu 
ciLower = propDiffWhiteBlack + cil
print('With 95% confidence interval for the proportion of white sounding : proportion of black sounding is {:.4} to {:.4}'.format(
    ciUpper,ciLower))
```

### Bootstrapping

```python
w = data[data.race=='w']
b = data[data.race=='b']
```

### Frequentist statistical approaches.

```python
# Define permutation functions:

def permutation_sample(data1, data2):
    """Return a permutation sample from two data sets."""

    # Concatenate the data sets:
    data = np.concatenate((data1, data2))

    # Permute the concatenated array:
    permuted_data = np.random.permutation(data)

    # Split the permuted array into two samples:
    permutation_sample_1 = permuted_data[:len(data1)]
    permutation_sample_2 = permuted_data[len(data1):]

    return permutation_sample_1, permutation_sample_2


def draw_permutation_replicates(data_1, data_2, func, size=1):
    """Draw multiple permutation replicates."""

    # Initialize array of permutation replicates:
    permutation_replicates = np.empty(size)

    for i in range(size):
        # Generate permutation samples:
        permutation_sample_1, permutation_sample_2 = permutation_sample(data_1, data_2)

        # Compute the test statistic:
        permutation_replicates[i] = func(permutation_sample_1, permutation_sample_2)

    return permutation_replicates


def difference_of_calls(data_1, data_2):
    """Return the difference in means of two arrays."""

    # The difference of means of data_1, data_2:
    return np.sum(data_1) - np.sum(data_2)

def diff_of_call_proportions(data_1, data_2):
    """Return the difference in means of two arrays."""

    # The difference of means of data_1, data_2:
    return np.sum(data_1)/len(data_1)-np.sum(data_2)/len(data_2)
```

```python
empirical_obs = diff_of_call_proportions(w.call, b.call)
print('Our observed data showed {:.5} difference in call backs'.format(empirical_obs))

perm_reps = draw_permutation_replicates(w.call, b.call, diff_of_call_proportions, 100000)

fig, ax = plt.subplots()
ax.hist(perm_reps, density=True, bins=24)
ax.axvline(empirical_obs, color='red')
ax.set_xlabel('Test-statistic (diff in proportion of call backs)')
ax.set_ylabel('PDF')
plt.show()
```

```python
std_err = np.std(perm_reps)
print('The standard error of our bootstrapped data is {:.5}'.format(std_err))

crit_z = stats.norm.ppf(0.95)
print('\ncrit_z at 1 tail 95% confidence: {:.4}'.format(crit_z))

moe = crit_z*std_err
print('The margin of error at 95% confidence: {:.4}'.format(moe))


[ciu,cil] = moe*np.array([-1,1])
ciUpper = 0 + ciu 
ciLower = 0 + cil
# Compute p-value: p
p = np.sum(perm_reps >= empirical_obs) / len(perm_reps)
print('p-value = {:.5f}\n'.format(p))
print('{} bootstraped replications had as or more extreme a result than our observed result'.format(np.sum(perm_reps >= empirical_obs)))
print('\nIf H0 is true and there is no difference between the populations with 95% confidence we expect: {} +/- {:.4}'.format(
    0, moe))
print('With 95% confidence interval for the proportion of white sounding : proportion of black sounding is {:.4} to {:.4}'.format(ciUpper,ciLower))
```
